### 7)   **Moving average**.

> The moving average (MA) is a simple technical analysis tool that smooths out price data by creating a constantly updated average price. The average is taken over a specific period of time, like 10 days, 20 minutes, 30 weeks or any time period the trader chooses.



```python
gm['Open'].plot(label = 'Without moving average', figsize = (15, 7))
gm['MA50'] = gm['Open'].rolling(50).mean( ) #MA50 means  moving average over 50 days
gm['MA50'].plot(label = 'MA50')
gm['MA200'] = gm['Open'].rolling(200).mean( ) #MA200 means  moving average over 200 days
gm['MA200'].plot(label = 'MA200')
plt.legend( )
```

# Daily percentage change


> r(t) = ( p(t)/ p(t-1) ) - 1

Here ->


1.   t is some day
2.   p(t) and p(t-1) is price of stock close on different days.








```python
tesla['returns'] = (tesla['Close'] / tesla['Close'].shift( 1 )) - 1
ford['returns'] = (ford['Close'] / ford['Close'].shift( 1 )) - 1
gm['returns'] = (gm['Close'] / gm['Close'].shift( 1 )) - 1
```

# Cumulative Return

```python
tesla['Cumulative Return'] = (1 + tesla['returns']).cumprod()
tesla.head()
```

```python
ford['Cumulative Return'] = (1 + ford['returns']).cumprod()
ford.head()
```

```python
gm['Cumulative Return'] = (1 + gm['returns']).cumprod()
gm.head()
```

```python
tesla['Cumulative Return'].plot( label = 'Tesla', figsize = (13, 6))
ford['Cumulative Return'].plot( label = 'Ford')
gm['Cumulative Return'].plot( label = 'Gm motors')
plt.legend()
plt.title('Correlation v.s. Time')
```

# Prediction for Tesla , Ford and GM motors for next 30 days using Linear Regression

### 1. Prediction for Tesla for next 30 days

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df_tesla = tesla
df_tesla.head( )
```

```python
df_tesla = df_tesla[['Adj Close']]
df_tesla.head( )
```

```python
# a variable for predicting n days into future
forecast_out = 30
df_tesla['Prediction'] = df_tesla[['Adj Close']].shift(-forecast_out)
df_tesla.head()
```

```python
#creating independent dataset   X
#converting dataframe to numpy
X = np.array(df_tesla.drop(['Prediction'], 1))
X = X[ :-forecast_out]
X
```

```python
#creating dependent dataset y
# converting dataframe to numpy
y = np.array(df_tesla['Prediction'])
y = y[ : -forecast_out]
y
```

```python
#split the data into 80% training and 20% testing
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
#create and train linear regression model
lr = LinearRegression( )
lr.fit(x_train, y_train)
```

```python
lr_confidence = lr.score(x_test, y_test)
print("Linear Regression Confidence for Tesla", lr_confidence)
```

```python
#set x_forecast to last 30 rows of original data from Adj Close Column
x_forecast = np.array(df_tesla.drop(['Prediction'], 1))[-forecast_out : ]
x_forecast
```

```python
lr_prediction_tesla = lr.predict(x_forecast)
lr_prediction_tesla
```

```python
df_tesla = pd.DataFrame(data = lr_prediction_tesla, columns= ["Prediction For Tesla"])
df_tesla.plot(label = 'Tesla', figsize=(13,7))
plt.legend()
plt.title(" Prediction of Tesla for next 30 days using Linear Regression Model")
```
